```python
# Preprocess the data for regression model
# Train a Regression Model
# Import the required libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from sentence_transformers import SentenceTransformer
import pickle
import re
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
from sklearn.preprocessing import LabelEncoder

# --- Preprocessing (with saving/loading and error handling) ---
embeddings_file = 'medication_embeddings.pkl'

try:
    # Load the embeddings, ensuring it's a DataFrame with correct length
    with open(embeddings_file, 'rb') as f:
        loaded_data = pickle.load(f)

    # Check if the loaded data is compatible with the DataFrame
    if isinstance(loaded_data, pd.DataFrame) and loaded_data.shape[0] == df.shape[0]:
        # Reset index to ensure alignment
        df = df.reset_index(drop=True)
        loaded_data = loaded_data.reset_index(drop=True)
        df['side_effects_embeddings'] = loaded_data['side_effects_embeddings']
    else:
        raise ValueError("Loaded embeddings are not compatible with the DataFrame.")

    print(f"Loaded pre-computed embeddings from {embeddings_file}.")
except (FileNotFoundError, EOFError, ImportError, ValueError) as e:
    print(f"Could not load embeddings from {embeddings_file}: {e}. Computing embeddings...")
    model = SentenceTransformer('all-mpnet-base-v2')

    def clean_html_robust(text):
        """Robustly cleans HTML and other unwanted characters."""
        try:
            # Remove HTML tags using BeautifulSoup
            soup = BeautifulSoup(text, 'html.parser')
            text = soup.get_text(separator=' ', strip=True)

            # Remove non-breaking spaces and other special characters
            text = re.sub(r'\xa0', ' ', text)  # Remove non-breaking spaces
            text = re.sub(r'[^\x00-\x7F]+', '', text)  # Remove non-ASCII characters
            text = re.sub(r'[\x00-\x1f]', '', text)  # Remove control characters
            text = text.strip()
            return text
        except Exception as e:
            print(f"Error cleaning text: {e}")
            return "Error cleaning text"

    # Print original data
    print("Original data:")
    print(df['side effects description'].head(20))

    df['side effects description'] = df['side effects description'].astype(str).apply(clean_html_robust)

    embeddings = model.encode(df['side effects description'].tolist())
    df['side_effects_embeddings'] = list(embeddings)

    # Save as a DataFrame with index reset
    with open(embeddings_file, 'wb') as f:
        pickle.dump(df[['side_effects_embeddings']].reset_index(drop=True), f)
    print(f"Embeddings computed and saved to {embeddings_file}.")

# Handle Missing Values
df.fillna("Unknown", inplace=True)

# --- Preprocessing for regression ---
side_effect_categories = ['renal', 'infection', 'dyslipidemia', 'other']

def assign_category(description):
    for category in side_effect_categories:
        if category in description.lower():
            return category
    return 'other'

df['side_effect_category'] = df['side effects description'].apply(assign_category)

label_encoder = LabelEncoder()
df['target_variable'] = label_encoder.fit_transform(df['side_effect_category'])

# --- Feature engineering and model training ---
X = df[['drug class', 'indication', 'therapeutic class description',
        'pharmacology description', 'dosage description']]
y = df['target_variable']

vectorizer = TfidfVectorizer()
text_columns = ['drug class', 'indication', 'therapeutic class description',
                'pharmacology description', 'dosage description']
tfidf_features = []
for col in text_columns:
    tfidf_features.append(vectorizer.fit_transform(X[col].astype(str)))

X_tfidf = hstack(tfidf_features)

X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

model = GradientBoostingRegressor()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Use appropriate regression metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
```

Interpretation:
MSE and MAE are both low, suggesting that the model predictions are reasonably close to the actual values.
An R-squared value of 0.0164 indicates that the model explains only about 1.64% of the variance in the target variable. This suggests a weak model performance, and I may need to explore feature engineering, hyperparameter tuning, or using different model architectures to improve my results.

```python
# Implement hyperparameter tuning using GridSearchCV for the GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV

# Define the model
model = GradientBoostingRegressor()

# Set up the parameter distribution for hyperparameter tuning
param_dist = {
    'n_estimators': [100, 150, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5, 6],
    'min_samples_split': [2, 3, 5],
    'min_samples_leaf': [1, 2, 3]
}

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist,
                                   scoring='neg_mean_squared_error', n_iter=20, cv=5,
                                   verbose=2, n_jobs=-1, random_state=42)

# Fit the model
random_search.fit(X_train, y_train)

# Get the best parameters and best score
best_params_random = random_search.best_params_
best_score_random = -random_search.best_score_

print(f"Best Parameters (Random Search): {best_params_random}")
print(f"Best Mean Squared Error from CV (Random Search): {best_score_random}")

# Train the model with the best parameters
best_model_random = GradientBoostingRegressor(**best_params_random)
best_model_random.fit(X_train, y_train)

# Evaluate the model on the test set
y_pred_best_random = best_model_random.predict(X_test)

# Calculate metrics
mse_best_random = mean_squared_error(y_test, y_pred_best_random)
mae_best_random = mean_absolute_error(y_test, y_pred_best_random)
r2_best_random = r2_score(y_test, y_pred_best_random)

print(f"Mean Squared Error (Best Model - Random Search): {mse_best_random}")
print(f"Mean Absolute Error (Best Model - Random Search): {mae_best_random}")
print(f"R-squared (Best Model - Random Search): {r2_best_random}")
```

Explanation:
GridSearchCV: This class performs exhaustive search over specified parameter values for an estimator. The best parameters are determined based on the cross-validated performance.
Parameter Grid: This grid specifies the values for different hyperparameters that I want to test.
Model Training and Evaluation: After finding the best parameters, I will train the model again and evaluate its performance.

RandomizedSearchCV: This class performs a random search over parameter values instead of an exhaustive search, which can significantly reduce the computation time.
n_iter: This parameter specifies the number of different combinations to try. Adjust this number based on time constraints.

```python
# Visaualize with Hierarchial Clustering
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans
import seaborn as sns

# Assuming you have a DataFrame `df` with relevant features
# Here we will use the 'side_effects_embeddings' for clustering visualization

# Convert embeddings to a suitable format for clustering
embeddings = np.array(df['side_effects_embeddings'].tolist())

# Select a subset of the data for visualization
subset_embeddings = embeddings[:30]  # Adjust as needed

# Perform hierarchical clustering
linked = linkage(subset_embeddings, method='ward')

# Apply KMeans to identify clusters (e.g., 3 clusters)
n_clusters = 3  # Define the number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(subset_embeddings)

# Create a color palette
palette = sns.color_palette("husl", n_clusters)  # Use different colors for each cluster

# Create a figure with a larger size
plt.figure(figsize=(14, 10))

# Create a dendrogram
dendro = dendrogram(
    linked,
    orientation='top',
    labels=[f'Sample {i+1}' for i in range(len(cluster_labels))],  # Simple labels
    color_threshold=0,  # Keep this at 0 to allow for custom coloring
)

# Create a color mapping for the lines in the dendrogram
for i, d in enumerate(dendro['dcoord']):
    # Determine the cluster index based on the leaves
    left_idx = int(dendro['ivl'][i].split()[1]) - 1
    right_idx = int(dendro['ivl'][i + 1].split()[1]) - 1 if i + 1 < len(dendro['ivl']) else left_idx

    # Get the cluster colors based on the labels
    left_color = palette[cluster_labels[left_idx]]
    right_color = palette[cluster_labels[right_idx]]

    # Set the color for the line
    plt.plot(d[0:2], [d[2], d[2]], color=left_color, linewidth=2)
    plt.plot(d[0:2], [d[2], d[2]], color=right_color, linewidth=2)

# Customize title and labels
plt.title('Hierarchical Clustering Dendrogram with Cluster Colors', fontsize=16)
plt.xlabel('Samples', fontsize=12)
plt.ylabel('Distance', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Integrate into Gradio Interface
import pickle
import pandas as pd
import numpy as np
import gradio as gr
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load the pre-trained components
with open("medication_embeddings.pkl", 'rb') as f:
    medication_embeddings = pickle.load(f)

# Check if medication_embeddings is a 2D numpy array
if not isinstance(medication_embeddings, np.ndarray) or medication_embeddings.ndim != 2:
    raise ValueError("medication_embeddings should be a 2D numpy array.")

# Load the SentenceTransformer model
sentence_model = SentenceTransformer('paraphrase-mpnet-base-v2')  # Ensure this is the same model used for generating embeddings

# Example DataFrame for medication details
df = pd.DataFrame({
    'generic name': ['Aspirin', 'Ibuprofen', 'Paracetamol', 'Amlodipine', 'Azithromycin'],
    'description': [
        'Pain reliever',
        'Anti-inflammatory',
        'Fever reducer',
        'Blood pressure medication',
        'Antibiotic drug used to treat infections'
    ],
    'side effects': [
        'Nausea, vomiting',
        'Stomach pain, headache',
        'Liver damage if overdosed',
        'Swelling, dizziness',
        'Nausea, diarrhea, abdominal pain'
    ]
})

def get_medication_info(medication_name):
    if not medication_name:
        return "Please enter a medication name."

    try:
        # Generate the embedding for the input medication name
        query_embedding = sentence_model.encode(medication_name)

        # Ensure query_embedding is a 1D array
        if query_embedding.ndim != 1:
            return "Error: Query embedding is not a 1D array."

        # Normalize the query embedding
        query_embedding = query_embedding / np.linalg.norm(query_embedding)

        # Normalize the medication embeddings
        normalized_medication_embeddings = medication_embeddings / np.linalg.norm(medication_embeddings, axis=1, keepdims=True)

        # Calculate cosine similarities between the normalized query embedding and normalized medication embeddings
        similarities = cosine_similarity(query_embedding.reshape(1, -1), normalized_medication_embeddings)

        # Get the index of the most similar medication
        most_similar_index = np.argmax(similarities)
        most_similar = df['generic name'].iloc[most_similar_index]

        # Debugging output for similarities
        print(f"Input: {medication_name}")
        print(f"Similarities: {similarities}")

        if most_similar in df['generic name'].values:
            matching_row = df[df['generic name'] == most_similar].iloc[0]
            info = matching_row.to_dict()  # Convert to dictionary for easier handling

            # Format output
            output_text = f"Information for {most_similar}:\n\n"
            for k, v in info.items():
                output_text += f"{k.replace('_', ' ').title()}: {v}\n"
            return output_text
        else:
            return "Medication not found."
    except Exception as e:
        return f"An error occurred: {str(e)}"

# Create the Gradio Interface
iface = gr.Interface(
    fn=get_medication_info,
    inputs=gr.Textbox(lines=2, placeholder="Enter medication name"),
    outputs="text",
    title="Medication Information Retrieval",
    description="Enter a medication name to get information."
)

# Launch the interface
iface.launch()
```
